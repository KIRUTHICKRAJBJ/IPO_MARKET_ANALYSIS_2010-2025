# Column-by-Column Explanation

## 1. Date

The listing date of the IPO.

Example:
2025-08-06 → The IPO listed on August 6, 2025.

## 2. IPO_Name

Name of the company that launched the IPO.

Examples:

M & B Engineering Ltd

Sri Lotus Developers & Realty Ltd

## 3. Issue_Size (crores)

Total amount the company aimed to raise from the IPO, in crores.

Examples:

₹650 crore

₹792 crore

Large issue size = bigger IPO = more institutional participation.

Investor Subscription Data

These columns show how much each category subscribed compared to the shares reserved for them.

## 4. QIB (Qualified Institutional Buyers)

How much institutional investors subscribed (mutual funds, banks, FPIs).

Example:
36.72 → QIBs subscribed 36.72 times their quota.

High QIB subscription usually means strong confidence from big institutions.

## 5. HNI (High Net-Worth Individuals)

Subscription from wealthy individuals.

38.24 → HNIs subscribed 38.24 times their reserved portion.

## 6. RII (Retail Individual Investors)

Subscription by normal retail investors.

32.55 → Retail was subscribed 32.55 times.

## 7. Total

Overall subscription across all categories.

Example:
36.20 → Overall IPO subscription was 36.20 times.

Higher = more demand → often leads to strong listing gains.

Listing Price & Gains
## 8. Offer Price

The price at which shares were offered during the IPO.

Examples:

₹385

₹150

## 9. List Price

The price at which the stock opened on listing day.

Example:

386.0 → Listed very close to offer price (flat listing)

179.1 → Strong premium over ₹150

## 10. Listing Gain (%)

Percentage gain (or loss) from Offer Price → List Price.

Example:

0.26% → Almost flat listing

19.40% → Good listing premium

Formula:

Listing Gain = ((List Price – Offer Price) / Offer Price) × 100

Current Market Performance
## 11. CMP (BSE)

Current share price on BSE.

## 12. CMP (NSE)

Current share price on NSE.

(Slight difference due to market dynamics.)

## 13. Current Gains (%)

How much the stock has gained from Offer Price till today.

Example:

10.87% → The stock grew 10.87% after listing.

34.07% → Strong post-listing performance.

Formula:

Current Gains = ((Current Price – Offer Price) / Offer Price) × 100

```python
# Extract Year from the Date 
ipo['Date_y'] = pd.to_datetime(ipo['Date'], format= '%y.%m.%d')
ipo['Year'] = ipo['Date_y'].dt.year
ipo['Year']
```

```python
# Count the number of IPO's by Year
ipo.groupby('Year')['IPO_Name'].count()
```

```python
# Total amount by Year
ipo.groupby('Year')['Total'].sum()
```

```python
# Show the IPO Company name where the List Price Greater than the Offer Price.
ipo['list_price_diff'] = ipo['List Price'] - ipo['Offer Price']
ipo[ipo['Offer Price'] < ipo['List Price']][['IPO_Name', 'Offer Price', 'List Price','list_price_diff']]
```

```python
# Issue size greaterthan 1000 cr
ipo[ipo['Issue_Size(crores)'] > 1000]['IPO_Name']
```

```python
sns.histplot(ipo,x='Year')
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(ipo,x='Year')
plt.ylabel("no.of.IPO's")
plt.title("Number Of IPO's by Year")
plt.show()
```

```python
# Aggregate the data by summing the 'Total' column per 'Year' ---
total_raised_per_year = ipo.groupby('Year')['Total'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.lineplot(data=total_raised_per_year, x='Year', y='Total', marker='o')
plt.ylabel("Total Money Raised (Currency Units)")
plt.xlabel('Year')
plt.title("Total Funds Raised by IPOs per Year")
plt.show()

```

```python
#we are trying to find the relation between issue_size and listing gain
plt.figure(figsize=(10, 5))
plt.scatter(ipo["Issue_Size(crores)"],ipo["Listing Gain"])
plt.xlabel("Issue Size")
plt.title("Box Plot of Issue Sizes")
plt.show()
```

```python
# relation between offer price and list price
plt.figure(figsize=(10, 5))
plt.scatter(ipo["Offer Price"],ipo["List Price"], color = "purple")
plt.xlabel("Issue Size")
plt.title("Box Plot of Issue Sizes")
plt.show()
```

```python
#top 10 ipo by listing gain

plt.figure(figsize=(10, 5))
top10 = ipo.nlargest(10, "Listing Gain")[["IPO_Name", "Listing Gain"]]
top10 = top10.iloc[::-1] # Reverse the order of the DataFrame
plt.barh(top10["IPO_Name"], top10["Listing Gain"])
plt.xlabel("Listing Gain")
plt.title("Top 10 IPOs by Listing Gain")
plt.show()
```
